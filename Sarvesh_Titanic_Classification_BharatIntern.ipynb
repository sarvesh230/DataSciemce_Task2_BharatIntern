```python
df= df.drop(columns=["PassengerId","Ticket","Fare","Name","Cabin"],axis=1)
```

```python
sns.pairplot(df,hue="Survived")
```

### Missing Values Treatment

```python
median1 = df["Age"].median()
df["Age"] = df["Age"].fillna(median1)
```

```python
df.isnull().sum()
```

```python
mode1 =df["Embarked"].mode()[0]
df["Embarked"] = df["Embarked"].fillna(mode1)
```

```python
df.isnull().sum()
```

### Dummies values

```python
# So that all the values gets converted into numeric forms
```

```python
df=pd.get_dummies(df,columns=["Sex","Embarked"])
```

```python
df.dtypes
```

### Developing the model by Logistic Regression

```python
X=df.drop(["Survived"],axis=1)
```

```python
y=df[['Survived']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
## Using Logistic because the Survived dependent variable is categorical value
model = LogisticRegression(solver='lbfgs')
model.fit(X_train,y_train)
model
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
cm=metrics.confusion_matrix(y_test,prediction,labels=[1,0])
df_cm=pd.DataFrame(cm,index=[i for i in["1","0"]],
                     columns=[i for i in ["Predict 1 ","Predict 0"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g',cmap="YlGnBu")
```
